```python
import pandas as pd

# Carregar o arquivo de voos novamente com o separador correto
df_voos_final = pd.read_csv(r"C:\Users\brend\OneDrive\Documentos\PYTHON\REPROGRAMA\PROJETO FINAL\Atividade\voos_final_atualizado.csv", sep=',', encoding='latin-1')

# Converter as colunas de data/hora para o tipo datetime com o formato correto
df_voos_final['Partida_Prevista'] = pd.to_datetime(df_voos_final['Partida_Prevista'], format='%d/%m/%Y %H:%M', errors='coerce')
df_voos_final['Partida_Real'] = pd.to_datetime(df_voos_final['Partida_Real'], format='%d/%m/%Y %H:%M', errors='coerce')

# Verificando quantos valores NaN foram introduzidos
print("\nNúmero de NaN em Partida_Prevista:", df_voos_final['Partida_Prevista'].isna().sum())
print("Número de NaN em Partida_Real:", df_voos_final['Partida_Real'].isna().sum())

```

```python
# 1. Convertendo as colunas de data/hora para o tipo datetime
df_voos_final['Partida_Prevista'] = pd.to_datetime(df_voos_final['Partida_Prevista'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
df_voos_final['Partida_Real'] = pd.to_datetime(df_voos_final['Partida_Real'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# 2. Calculando o atraso (em minutos)
df_voos_final['Atraso'] = (df_voos_final['Partida_Real'] - df_voos_final['Partida_Prevista']).dt.total_seconds() / 60

# 3. Remover registros com atrasos NaN
df_voos_final = df_voos_final.dropna(subset=['Atraso'])

# 4. Agrupar pelos aeroportos de origem e calcular a média de atrasos
media_atrasos = df_voos_final.groupby('Sigla_Aeroporto_Origem')['Atraso'].mean().sort_values(ascending=False).head(5)

# **DEBUG: Imprimir média de atrasos**
print("Média de Atrasos por Aeroporto de Origem:")
print(media_atrasos)

# 5. Criar o gráfico apenas se houver dados
if not media_atrasos.empty:
    plt.figure(figsize=(12, 8))
    sns.barplot(x=media_atrasos.index, y=media_atrasos.values, palette='coolwarm')  # Usando a paleta Coolwarm
    plt.title('Média de Atrasos por Aeroporto de Origem', fontsize=16, fontweight='bold')
    plt.xlabel('Aeroporto de Origem', fontsize=14)
    plt.ylabel('Média de Atraso (minutos)', fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.yticks(fontsize=12)

    # Adicionando a grade
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Adicionando anotações nas barras
    for i, v in enumerate(media_atrasos.values):
        plt.text(i, v + 1, f"{v:.1f}", ha='center', fontsize=10, color='black')  # Ajusta a posição da anotação

    plt.tight_layout()  # Ajusta o layout para evitar sobreposição
    plt.show()
else:
    print("Nenhum dado disponível para plotar o gráfico.")
```
